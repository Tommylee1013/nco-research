```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns

data = yf.download(
    [
        '^GSPC',
        '^GSPE',
        '^SP500-15',
        '^SP500-20',
        '^SP500-25',
        '^SP500-30',
        '^SP500-35',
        '^SP500-40',
        '^SP500-45',
        '^SP500-50',
        '^SP500-55',
        '^SP500-60',
    ],
    start = '2020-01-01',
    interval = '1d',
    progress = False,
    multi_level_index = False,
    auto_adjust = False
)
```

```python
sector_ret = data['Close'].pct_change(fill_method = None)
```

```python
sns.clustermap(
    sector_ret.corr(), 
    figsize = (6,6)
)
```
